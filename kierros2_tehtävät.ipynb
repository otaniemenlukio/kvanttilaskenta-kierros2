### Tiivistelmä kompleksiluvuista
Kompleksiluku on muotoa $z=a+bi$, missä $i$ on _imaginaariyksikkö_ eli luku, jolla on ominaisuus $i^2=-1$.

Luku $a$ on kompleksiluvun reaaliosa ja luku $b$ imaginaariosa. 

Erilaisia kompleksilukuja ovat esimerkiksi $3+2i$, $-5i$ ja $13$.

Kompleksiluvun $z=a+bi$ _liittoluku_ eli _kompleksikonjugaatti_ on $\overline{z}=a-bi$.

Kompleksilukuja esitetään ns. kompleksitasossa, jossa kompleksilukua z havainnollistetaan tason pisteenä $(a,b)$. Kompleksiluvun itseisarvo eli _moduli_ ilmaisee tällöin kompleksiluvun paikkavektorin pituuden

$$
\left|z\right|=\left|a+bi\right|=\sqrt{a^2+b^2}
$$

## Tehtävä 5 matriisin transponointi ja kompleksikonjugaatti

Kaksiulotteinen matriisi $A$: 

$$
A=\left(\begin{matrix}
a&b\\
c&d
\end{matrix}\right)\ 
$$

Matriisin $A$ transpoosi $A^T$ määritellään seuraavasti:
$$
A^T=\left(\begin{matrix}
a&c\\
b&d
\end{matrix}\right)\ 
$$

Matriisin $A$ kompleksikonjugaatti $\overline A$ muodostetaan laskemalla jokaisen alkion liittoluku eli konjugaatti. Esimerkiksi:

$$
A=\left(\begin{matrix}
i&3\\
2&1-2i
\end{matrix}\right)
\\
\overline A=\left(\begin{matrix}
-i&3\\
2&1+2i
\end{matrix}\right)\ 
$$

Matriisin $A$ konjugaattitranspoosi eli $\overline{A^T}$ merkitään kvanttimekaniikassa $A^\dagger$: 

$$
A^\dagger=\left(\begin{matrix}
-i&2\\
3&1+2i
\end{matrix}\right)\ 
$$

Pythonissa matriisin konjukaattitraspoosin saa kirjoittamalla `.H` (tulee sanasta hermiittinen konjugaatti). Esimerkiksi:

```python
zero = np.matrix([[1], [0]])
zero.H
```
